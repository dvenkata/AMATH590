```python
X = np.load('/content/drive/MyDrive/Colab Notebooks/vel3_cutout.npy')
X.shape
```

```python
nx1 = round(X.shape[0])
ny1 = round(X.shape[1])
nz1 = round(X.shape[2])
tt1 = X.shape[4]
Y = np.zeros((nx1,ny1,nz1,tt1))
```

```python
#Compute Energy
u = np.zeros((X.shape[0],X.shape[1],X.shape[2],1,X.shape[4]))
u[:,:,:,0,:] = X[:,:,:,0,:]**2 + X[:,:,:,1,:]**2 + X[:,:,:,2,:]**2
```

```python
Y = u[:,:,:,0,:]
```

```python
#Plot a surface at a given z, over time
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (15,15))
a = fig.add_subplot(121)
a.title.set_text('t = 1')
a.imshow(Y[:,:,0,0], cmap='jet')

a = fig.add_subplot(122)
a.title.set_text('t = 257')
a.imshow(Y[:,:,0,1], cmap='jet')
```
